```python
#
%matplotlib inline
%config InlineBackend.figure_format = "retina"
from IPython.core.pylabtools import figsize
import numpy as np
import matplotlib.pyplot as plt

#(a)   
def beat(A, B, fc, df, dur, fs = 11025):
    t = np.arange(0, dur, 1/fs)
    x1 = A*np.exp(-0.0007*2*np.pi*(fc-df)*t)*np.cos(2*np.pi*(fc-df)*t)
    x2 = B*np.exp(-0.0007*2*np.pi*(fc+df)*t)*np.cos(2*np.pi*(fc+df)*t)
    
    x4 = B*np.exp(-0.0007*2*np.pi*(319.5+2.5)*t)*np.cos(2*np.pi*(319.5+2.5)*t)
    x = x1 + x2 + x4
    return x, t

#(b)    
from scipy import signal
fc, delf, fs, dur = 168, 5, 11025, 3
x, t = beat(30, 30, fc, delf, dur)
fig, ax = plt.subplots(figsize=(6,2.5))
ax.plot(t*1000, x)
ax.set_xlim(0, 2000)
ax.set_yticks([-1, 0, 1])
ax.set_xlabel('t(msec)')
ax.set_title('beat signal')
fig.tight_layout()

plt.show()

#(c)  
from IPython.display import Audio
Audio(data = x, rate = fs)
```

```python
#2
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
from IPython.core.pylabtools import figsize
import numpy as np
import matplotlib.pyplot as plt

#   
def beat(A, B, fc, df, dur, fs=11025):
    t = np.arange(0, dur, 1/fs)
    x1 = 0.01*np.exp(-21.1*0*t)*np.cos(2*np.pi*(11.0)*t)
    x2 = 0.01*np.exp(-33.8*0*t)*np.cos(2*np.pi*(313)*t)
    x3 = 0.1*np.exp(-14.1*0*t)*np.cos(2*np.pi*(622)*t)
    x4 = 1*np.exp(0.0*0*t)*np.cos(2*np.pi*(934)*t)
    x5 = 0.3*np.exp(-8.7*0*t)*np.cos(2*np.pi*(1246)*t)
    x6 = 0.25*np.exp(-8.9*0*t)*np.cos(2*np.pi*(1556)*t)
    x7 = 0.5*np.exp(-6.7*0*t)*np.cos(2*np.pi*(1867)*t)
    x8 = 0.01*np.exp(-19.7*0*t)*np.cos(2*np.pi*(2179)*t)
    x9 = 0.01*np.exp(-24.7*0*t)*np.cos(2*np.pi*(2488)*t)
    x10 = 0.01*np.exp(-29.6*0*t)*np.cos(2*np.pi*(2800)*t)
    x11 = 0.01*np.exp(-34.4*0*t)*np.cos(2*np.pi*(3111)*t)

    x = x1 + x2 + x3 + x4 + x5 + x6 + x7 + x8 + x9 + x10 + x11
    return x, t

#     
from scipy import signal
fc, delf, fs, dur= 440, 5, 11025, 3
x, t = beat(1, 1, fc, delf, dur)

fig, ax = plt.subplots(figsize=(6,2.5))
ax.plot(1000*t, x)
ax.set_xlim(0, 200)
ax.set_xlabel([-1, 0, 1])
ax.set_xlabel('t(msec)')
ax.set_title('beat signal')
fig.tight_layout()
plt.show()

#  
from IPython.display import Audio
Audio(data=x, rate=fs)
```

```python
# (d)    
from numpy.fft import fft, fftshift

Nfft = 2048            # FFT  
X = fftshift(fft(x, Nfft) / fs)   #  Nfft  FFT, fs 
f = np.linspace(-fs/2, fs/2, Nfft, endpoint=False)

#   dB  ( 0 dB )
Xmag = np.abs(X)
Xmag_db = 20 * np.log10(Xmag / np.max(Xmag) + 1e-12)

#  (0~4000 Hz) 
idx = (f >= 0) & (f <= 4000)

fig, ax = plt.subplots(figsize=(6, 2.5))
ax.plot(f[idx], Xmag_db[idx], lw=2)
ax.set_xlim(0, 4000)
ax.set_ylim(-60, 0)          
ax.set_xlabel('f(Hz)')
ax.set_ylabel('Magnitude (dB)')
ax.set_title('spectrum of beat signal')
fig.tight_layout()
plt.show()
```
